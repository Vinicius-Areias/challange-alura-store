#1. Análise do faturamento


```python
import pandas as pd
```

```python
arquivos = ["loja_1.csv", "loja_2.csv", "loja_3.csv", "loja_4.csv"]
```

```python
df_list = []
```

```python
urls = ["https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"]

for url in urls:
    temp = pd.read_csv(url)
    temp["Loja"] = url.split("/")[-1].replace(".csv","")
    df_list.append(temp)
```

```python
df = pd.concat(df_list, ignore_index=True)
```

```python
print("Prévia dos dados combinados:")
display(df.head())
```

```python
print("Prévia dos dados combinados:")
```

```python
display(df.head())
```

```python
faturamento_por_loja = df.groupby("Loja")["Preço"].sum().reset_index()
```

```python
faturamento_por_loja.rename(columns={"Preço": "Faturamento_Total"}, inplace=True)
```

```python
print("Faturamento total por loja:")
```

```python
display(faturamento_por_loja)
```

```python
faturamento_total = df["Preço"].sum()
```

```python
print(f"\nFaturamento total (todas as lojas): R$ {faturamento_total:,.2f}")
```

```python
faturamento_ordenado = faturamento_por_loja.sort_values(by="Faturamento_Total", ascending=False)
```

```python
print("\nFaturamento das lojas em ordem decrescente:")
```

```python
display(faturamento_ordenado)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

```python
df = pd.concat([pd.read_csv(url).assign(Loja=url.split("/")[-1].replace(".csv","")) for url in urls], ignore_index=True)
```

```python
faturamento_por_loja = df.groupby("Loja")["Preço"].sum().reset_index(name="Faturamento_Total")
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=faturamento_por_loja, x="Loja", y="Faturamento_Total")
plt.title("Faturamento Total por Loja")
plt.ylabel("Faturamento (R$)")
plt.xlabel("Lojas")
plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
import pandas as pd
```

```python
arquivos = ["loja_1.csv", "loja_2.csv", "loja_3.csv", "loja_4.csv"]
```

```python
df_list = []
```

```python
for url in urls:
    temp = pd.read_csv(url)
    temp["Loja"] = url.split("/")[-1].replace(".csv", "")
    df_list.append(temp)
```

```python
df = pd.concat(df_list, ignore_index=True)
```

```python
print("Prévia dos dados combinados:")
```

```python
display(df.head())
```

```python
vendas_por_produto = df.groupby(["Loja", "Produto"]).size().reset_index(name="Quantidade_Vendas")
```

```python
print("Quantidade de vendas por produto em cada loja:")
```

```python
display(vendas_por_produto)
```

```python
mais_e_menos_vendidos = vendas_por_produto.groupby("Loja").apply(
    lambda x: pd.DataFrame({
        "Mais_Vendido": [x.loc[x["Quantidade_Vendas"].idxmax(), "Produto"]],
        "Qtd_Mais_Vendido": [x["Quantidade_Vendas"].max()],
        "Menos_Vendido": [x.loc[x["Quantidade_Vendas"].idxmin(), "Produto"]],
        "Qtd_Menos_Vendido": [x["Quantidade_Vendas"].min()]
    })
).reset_index(drop=True)
```

```python
print("Produtos mais e menos vendidos por loja:")
```

```python
display(mais_e_menos_vendidos)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

```python
arquivos = ["loja_1.csv","loja_2.csv","loja_3.csv","loja_4.csv"]
df = pd.concat([pd.read_csv(url).assign(Loja=url.split("/")[-1].replace(".csv","")) for url in urls], ignore_index=True)
```

```python
top10_produtos = (df.groupby("Produto")
                    .size()
                    .sort_values(ascending=False)
                    .head(10))
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top10_produtos.values, y=top10_produtos.index)
plt.title("Top 10 Produtos Mais Vendidos (Todas as Lojas)")
plt.xlabel("Quantidade de Vendas")
plt.ylabel("Produto")
plt.tight_layout()
plt.show()
```
